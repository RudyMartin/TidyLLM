```python
import boto3
import json

# Initialize Bedrock client
bedrock = boto3.client("bedrock")

def list_bedrock_models():
    """Fetch and display available Amazon Bedrock models."""
    try:
        response = bedrock.list_foundation_models()
        models = response.get("modelSummaries", [])

        print("\n🔍 Available Amazon Bedrock Models:")
        for model in models:
            print(f"🛠 Model ID: {model['modelId']}")
            print(f"   Provider: {model['providerName']}")
            print(f"   Output Modality: {model.get('outputModality', 'N/A')}")
            print(f"   Model ARN: {model['modelArn']}\n")

    except Exception as e:
        print(f"❌ Error retrieving Bedrock models: {e}")

# Run the function
list_bedrock_models()
```
